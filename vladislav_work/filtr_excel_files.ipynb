Найден файл, который содершит в себе пропущенные данные и их стоит заполнить.

```python
import pandas as pd

file = pd.read_excel('Макропоказатели.xlsx')
file.head()
```

Первый не заполненный столбик решил записать нулями

```python
file['Базовый индекс потребительских цен на товары и услуги'] = file['Базовый индекс потребительских цен на товары и услуги'].fillna(0)
file.head()
```

Второй не заполненный столбик решил заполнить средним значением

```python
mean_value = file['Ключевая ставка'].mean()

file['Ключевая ставка'] = file['Ключевая ставка'].fillna(mean_value)
file.head()
```

Третий столбик заполнил тем же способом

```python
mean_value = file['Курс доллара'].mean()

file['Курс доллара'] = file['Курс доллара'].fillna(mean_value)
file.head()
```
